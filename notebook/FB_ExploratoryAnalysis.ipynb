## Importing libraries for data analysis

```python
import pandas as pd
import numpy as np
import pickle
```

```python
filename = "../output/10207648390994439"
def readData(filename):
    with open(filename, "rb") as fileHandler:
        data = pickle.load(fileHandler)
        return data
data = readData(filename)
```
