# <center> Facebook Friend Recommendation </center>

## Importing libraries for data analysis

```python
import networkx as nx
```

## Reading the Univ of Washington Dataset

```python
G = nx.read_edgelist("../data/facebook-links.txt.anon", data=(("timestamp", str),))
```

## Displaying the edges of the graph

```python
G.edges(data=True)
```

## How many friends does each user has ?

```python
G.degree()
```

## What fractions of user friends are mutual friends?

```python
nx.clustering(G)
```
